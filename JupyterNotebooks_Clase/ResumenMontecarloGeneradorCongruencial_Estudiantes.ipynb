### **Simular el lanzamiento de una moneda**

Para el lanzamiento de la moneda necesitamos determinar sus probabilidades.


1.   Probabilidad Cara
2.   Probabilidad Sello
3.   Probabilidad Lomo
    
|Valor|Probabilidad|
|:-:|:-:|
|Probabilidad Cara|49.5%|
|Probabilidad Sello|49.5%|
|Probabilidad Lomo|01.0%|




```python
#Lanzamiento de moneda
#Crear algoritmo
import random as rnd
aleatorio = rnd.random() #Entrega valores entre 0 y 1
if aleatorio <= 0.495:
    print('Cara')
elif aleatorio <= 0.99:
    print('Sello')
else:
    print('Lomo')
```

```python
#Ahora en funcion
def fResultadoMonedaTexto():
    aleatorio = rnd.random()
    if aleatorio <= 0.495:
        return 'Cara'
    elif aleatorio <= 0.99:
        return 'Sello'
    else:
        return 'Lomo'
```

## Simular el siguiente ejercicio
### LA TIENDA

Luego de realizar una investigación por parte de la llegada al país de la empresa OnlyFans, han comisionado una investigación para ver donde pueden poner varias oficina de asesoría y han elegido tres ciudades detalladas a continuación con sus respectivas cantidades de oficinas. Realizar una simulación en donde las ciudades tienen la misma probabilidad y detallar donde posiblemente con una cantidad de repeticiones anidada de al menos 100 iteraciones se deben poner las oficinas de OnlyFans.

    
|Medellín|Bogotá|Cali|
|:-:|:-:|:-:|
|Tienda 1→0.25|Tienda 1→0.75|Tienda 1→0.333|
|Tienda 2→0.50|Tienda 2→0.25|Tienda 2→0.333
|Tienda 3→0.10||Tienda 3→0.334|
|Tienda 4→0.15|||


```python
n = 1000 #@param {type:"slider", min:1000, max:100000, step:1000}
experimentos = 100 #@param {type:"slider", min:100, max:1000, step:100}
ruta = ['Medellín', 'Bogotá', 'Cali']
#En Medallo
#Tienda 1 0.25, Tienda 2 0.5, Tienda 3 0.10, Tienda 4 0.15
#En Bogotá
#Tienda 1 0.75, Tienda 2 0.25
#En Cali
#Tienda 1 0.333, Tienda 2 0.333, Tienda 3 0.334

t1m, t2m, t3m, t4m, t1b, t2b, t1c,t2c,t3c = [],[],[],[],[],[],[],[],[]
tiendamed, tiendabog, tiendacali = [],[],[]
for i in range(experimentos):
    for j in range(n): #Vamos a realizar n experimentos
        x = rnd.choice(ruta)
        if x == 'Medellín':
            y = rnd.random()
            if y > 0.85:
                t4m.append(1)
            elif y > 0.75:
                t3m.append(1)
            elif y > 0.25:
                t2m.append(1)
            else:
                t1m.append(1)
        if x == 'Bogotá':
            y = rnd.random()
            if y > 0.75:
                t2b.append(1)
            else:
                t1b.append(1)
        if x == 'Cali':
            y = rnd.random()
            if y > 0.666:
                t3c.append(1)
            elif y > 0.333:
                t2c.append(1)
            else:
                t1c.append(1)
    #luego de tener mi arbol de decision debo tomar una decision
    sumamedellin = sum(t1m) + sum(t2m) + sum(t3m) + sum(t4m)
    sumabogota = sum(t1b) + sum(t2b)
    sumacali = sum(t1c) + sum(t2c) + sum(t3c)

    if sumamedellin > sumabogota and sumamedellin > sumacali:
        tiendamed.append(1)
    if sumabogota > sumamedellin and sumabogota > sumacali:
        tiendabog.append(1)
    if sumacali > sumamedellin and sumacali > sumabogota:
        tiendacali.append(1)
print('Cantidad de veces tienda Medellin {}'.format(sum(tiendamed)))
print('Cantidad de veces tienda Bogota {}'.format(sum(tiendabog)))
print('Cantidad de veces tienda Cali {}'.format(sum(tiendacali)))
```

## Simular un juego de Bingo sin repetir letra y numero generado. (Ya explicado en clase, se repetira por pertinencia)
### Bingo 75

Es el Bingo americano, muy jugado en red. Tiene formato de cartón y juego de 24 números sobre 75 (la casilla central no tiene número). Basado en la realización de figuras dentro del cartón se pueden obtener diferentes posibilidades de premios de bingo (líneas, diagonales, cruces, etc...). Estas figuras se llaman “patrones”, y cada juego tiene uno o más patrones que dan premio. La composición del cartón es la siguiente: en la columna “B” se ubican los números entre el 1 y el 15, en la columna “I” los números del 16 al 30, en la “N” los números del 31 al 45, en la “G” del 46 al 60 y finalmente en la columna “O” los números del 61 al 75. Los patrones para premios de Bingo como se ha dicho antes, pueden ser de cualquier forma: en diagonal, línea, cruces, cuadrados, letras... En cada juego se especifica qué patrón da premio.

![Bingo75](https://upload.wikimedia.org/wikipedia/commons/4/47/Cartonbingo.jpg)

El ejercicio consiste en crear un cartón de BINGO o en su defecto $n$ cartones utilizando una función que retorne una matriz.

```python
import numpy as np
#Necesitas numpy para crear e iterar matrices
#Debemos crear una funcion para crear n cartones
def fCrearCartonBINGO():
    #primero debemos crear las listas con los numeros para poblar la matriz cartonuda
    B = []
    I = []
    N = []
    G = []
    O = []
    X = []
    #Con varios ciclos, tambien se puede hacer con un solo ciclo y condiciones
    for i in range(1, 16):
        B.append(i)
    for i in range(16, 31):
        I.append(i)
    for i in range(31, 46):
        N.append(i)
    for i in range(46, 61):
        G.append(i)
    for i in range(61, 76):
        O.append(i)
    #ya tengo los valores, ahora debo recorrer una matriz de 6x5 con las letras BINGO en la primera fila
    A = np.zeros((6, 5), dtype='O') # Creamos una matrix de 6x5 de tipo txto

    #Ahora para recorrer la matrix debemos iterar los elementos
    #Escribir la palabra BINGO
    bingo = ['B','I', 'N', 'G', 'O']
    columnas = A.shape[1] # 0 es para filas, 1 para columnas
    filas = A.shape[0]
    for q in range(columnas):
        A[0,q] = bingo[q]

    for i in range(filas):
        for j in range(columnas):
            if i > 0:
                if j == 0: # Estamos en la columna 1, debemos llamar la B
                    dato =  rnd.choice(B)
                    B.remove(dato)
                    A[i,j] = dato
                elif j == 1:
                    dato =  rnd.choice(I)
                    I.remove(dato)
                    A[i,j] = dato
                elif j == 2:
                    dato =  rnd.choice(N)
                    N.remove(dato)
                    A[i,j] = dato
                elif j == 3:
                    dato =  rnd.choice(G)
                    G.remove(dato)
                    A[i,j] = dato
                else:
                    dato =  rnd.choice(O)
                    O.remove(dato)
                    A[i,j] = dato
            if i == 3 and j == 2:
                A[i,j] = '-'
    return A
```

```python
#@title Seleccionar la cantidad de cartones a generar
cartones = 10 #@param {type:"slider", min:1, max:10, step:1}
for i in range(cartones):
    print('Carton {}'.format(i+1))
    print(fCrearCartonBINGO())
    print('---------------------------')
```

### Dominó

El dominó es un juego de mesa en el que se juegan y emplean unas fichas (baldosas) rectangulares, generalmente blancas por la cara y negras por el envés, usualmente hay de diferentes colores, aunque existen muchas variantes. Una de sus caras está dividida por dos cuadrados, cada uno de los cuales está numerado normalmente mediante disposiciones de puntos como los dados.
La puntuación habitual es de cero a seis puntos, lo que compone un total de 28 piezas de dominó siendo la ficha más grande la del seis doble. Existen otras variantes de 55, 91, 136 y 190 piezas. Por ejemplo, en Cuba se juega con 28 piezas en la zona oriental, mientras que en el centro y occidente, incluyendo la capital, se juega con 55 fichas (Doble 9). El jugador que gana sin fichas son 3 puntos y por medio de puntos son 2. Si un jugador al de la partida obtiene cuatro fichas con iguales números en ambos lados es considerado como victorioso y la partida será reiniciada.

![Dominó](https://upload.wikimedia.org/wikipedia/commons/a/ae/Dominospiel.JPG)

#Generador Congruencial Final
## Requisitos
### 1.    $C <> 0$;  primo relativo de $M$
### 2.    $a \ Mod \ q = 1$;  q y a, deben ser primos relativos de $M$, $a$ debe ser primo
### 2.    $a \ Mod \ 4 = 1$;  4 debe ser primo relativo de $M$

El generador congruencial tiene la siguiente forma:

<h1>

$X_{i} = (a*X_{i-1}+C) Mod \ M$

</h1>

<h1>

$R_{i} = \frac{ X_{i}}{M}$

</h1>


```python
import math as mt

#Funciones para generar numeros primos y para generar una semilla
def NumerosPrimos(inferior, superior):
    primos = []
    for num in range(inferior,superior+1):
        if num > 1:
            for i in range(2,num):
                if (num%i) == 0:
                    break
            else:
                primos.append(num)
    return primos

# Generador de semillas
def Semilla():
    tiempo = datetime.now()
    sem = float(tiempo.strftime('.%f'))
    sem = round(sem,6)
    sem = int(sem*1000000)
    return sem
```
