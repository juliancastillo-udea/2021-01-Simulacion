```python
import matplotlib.pyplot as plt
import statistics as st
from datetime import datetime

def semilla_():
    tiempo = datetime.now()
    milisegundos = float(tiempo.strftime('.%f'))
    milisegundos = round(milisegundos, 3) # Si, se sabe el k
    milisegundos = int(milisegundos * 1000)
    if milisegundos < 100:
        milisegundos += 100 # Con recursividad - Programacion supersayayin
    return milisegundos

error = '0'
vsemillas, vresultados, valeatorios = [], [], []
iteraciones = int(input('Ingresar la cantidad de iteraciones: '))
k = len(str(semilla_()))
divisor = 10**k
semilla = semilla_()
j = 0
for i in range(iteraciones):
    resultado = semilla * semilla
    if resultado != 0:
        aleatorio = str(resultado)
        if len(aleatorio) > 5: #1234567
            aleatorio = aleatorio[1:4] # DC 1 2 3 
        elif len(aleatorio) > 4:
            aleatorio = aleatorio[1:3]
        elif len(aleatorio) > 3:
            aleatorio = aleatorio[1:2]
        else:
            aleatorio = aleatorio[1:]
        semilla = int(aleatorio)
        vsemillas.append(semilla)
        valeatorios.append(float(int(aleatorio)/divisor))
        vresultados.append(resultado)
    else:
        error = 'Oh, se ha presentado un error, Hemos encontrado un cero.'
        break
    if j > 1:
        if valeatorios[j-1] == valeatorios[j]:
            break
    j += 1
uni = 1/2
inde = 1/12
promedio = st.mean(valeatorios)
varianza = st.variance(valeatorios)
#Comparar promedio y uni, inde con varianza Si su criterio de aceptacion es menor a un alfa 1.96 acepta
print(promedio, varianza)
print(valeatorios)
```
