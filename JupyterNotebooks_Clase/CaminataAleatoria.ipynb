# **Caminata Aleatoria**
## *Para Norte, Sur, Este y Oeste*

```python
pasos = 30 #@param {type:"slider", min:0, max:100, step:10}
import numpy as np
import matplotlib. pyplot as plt
import random as rnd
x = np.zeros(pasos)
y = np.zeros(pasos)
for i in range(pasos):
    a = rnd.random()
    if a < 0.25:    #Norte
        x[i] = x[i-1]+0
        y[i] = y[i-1]+1
    elif a < 0.5:   #Este
        x[i] = x[i-1]+1
        y[i] = y[i-1]+0
    elif a < 0.75:  #Sur
        x[i] = x[i-1]+0
        y[i] = y[i-1]-1
    else:           #Oeste
        x[i] = x[i-1]-1
        y[i] = y[i-1]+0
print('los valores generados para Von Neumann son:')
print(x,y)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Caminata Aleatoria')
plt.grid('on')
```

## *Para el vecindario de Moore*

```python
pasos2 = 100 #@param {type:"slider", min:0, max:100, step:10}
```

```python
#Caminata aleatoria 8 direcciones
# ↖ ↗ ↘ ↙ ↓ → ↑ ←
# En sentido de las manecillass del Reloj
#  _______________
# |1/8 | 2/8 | 3/8|
# |8/8 | 1/1 | 4/8|
# |7/8 | 6/8 | 5/8|
#  ‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾
x = np.zeros(pasos2)
y = np.zeros(pasos2)
for i in range(pasos2):
    a = rnd.random()
    if a < 1/8:     # ↖
        x[i] = x[i-1] - 1
        y[i] = y[i-1] + 1
    elif a < 2/8:   # ↑
        x[i] = x[i-1] + 0
        y[i] = y[i-1] + 1
    elif a < 3/8:   # ↗
        x[i] = x[i-1] + 1
        y[i] = y[i-1] + 1
    elif a < 4/8:   # →
        x[i] = x[i-1] + 1
        y[i] = y[i-1] + 0
    elif a < 5/8:   # ↘
        x[i] = x[i-1] + 1
        y[i] = y[i-1] - 1
    elif a < 6/8:   # ↓
        x[i] = x[i-1] + 0
        y[i] = y[i-1] - 1
    elif a < 7/8:   # ↙
        x[i] = x[i-1] - 1
        y[i] = y[i-1] - 1
    else:           # ←
        x[i] = x[i-1] - 1
        y[i] = y[i-1] + 0
print('los valores generados para Moore son:')
print(x,y)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Caminata Aleatoria')
plt.grid('on')
```
