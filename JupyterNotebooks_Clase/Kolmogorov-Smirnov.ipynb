#Prueba de Kolmogorov-Smirnov

##Pasos
El procedimiento general para realizar esta prueba se puede resumir en los siguientes pasos: 



1.   Organizar los datos de menor a mayor, donde &𝑅_𝑖& denote la i-esima menor observación así:$𝑅_1≤𝑅_2≤…≤𝑅_𝑁$
2.   Calcular
>$𝐷^+ =  \smash{\displaystyle\max_{1 \leq i \leq N}} \left\{\frac{i}{N}-R_i \right\}$
>
>$𝐷^- =  \smash{\displaystyle\max_{1 \leq i \leq N}} \left\{R_i -\frac{i-1}{N} \right\}$
3.   Calcular
>$D=max(D^+, D^-)$
4.   Localizar el valor critico $D_\alpha$ en la tabla dado el valor de significancia y el tamaño de N.
5.   Si el tamaño de $D$ es mayor que el valor critico $D_\alpha$, la hipótesis nula de que los datos son una muestra de la distribución uniforme es rechazado. Si $D \leq D_\alpha$ se concluye que no hay diferencia detectada entre la distribución muestra y una distribución uniforme.




```python
import pandas as pd
import random as rnd
import numpy as np
#Cargamos la tabla de Chi-Cuadrado
filename = r"https://github.com/juliancastillo-udea/2021-01-Simulacion/blob/main/KolmogorovSmirnov.csv?raw=True"
data = pd.read_csv(filename, sep=';', usecols=["TamanioMuestra", "Alfa0.05"], index_col ='TamanioMuestra')
ks = data.loc[36] #Posicion 36 para mas de 35 registros
ks = ks[0]
#Generamos n numeros aleatorios
aleatorios = []
for i in range(100):
    aleatorios.append(rnd.random())
#Debemos definir i
i = len(aleatorios)
N = i
#Debemos generar un listado de la misma cantidad de datos entre 0 y 1 equidistribuidos
a = np.arange(0,1,1/i)
a = list(a)
aleatorios.sort() #Organizamos los valores aleatorios
#Recorremos los datos aleatorios y calculamos D+ y D-
dmas, dmenos = [], []
for x in range(len(aleatorios)):
    i=x
    Ri = aleatorios[x]
    #+++++++++++++++++++++++++++++++
    dma = (i/N)-Ri
    dmas.append(dma)
    #-------------------------------
    dme = Ri-((i-1)/N)
    dmenos.append(dme)
    ################################
D = [max(dmas), max(dmenos)]
q = max(D)
ks = ks/(N**(1/2))
if q <= ks:
    print('Aceptamos hipotesis que los numeros vienen de una distribucion uniforme')
else:
    print('Rechazamos hipotesis que los numeros vienen de una distribucion uniforme')
```
