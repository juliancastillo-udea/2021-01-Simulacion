## Introducción - Calcular Pi

![CircunferenciaUnitaria](https://www.disfrutalasmatematicas.com/geometria/images/unit-circle.svg)

## $\frac{\pi*r^2}{a*b} = \frac{\pi*1^2}{2*2} = \frac{\pi}{4}$

````
import random as rnd
ensayos = input('Ingresar el numero de ensayos: ')
hit = 0
contador = 0
ensayos = int(ensayos)
vxhit = []
vyhit = []
vxmiss = []
vymiss = []
#hit or miss - acierto o error
for z in range(ensayos): #range crea numeros enteros desde 0 hasta ensayo
    x = rnd.uniform(-1,1)
    y = rnd.uniform(-1,1)
    if (x**2+y**2) <= 1:
        hit += 1
        vxhit.append(x)
        vyhit.append(y)
    else:
        vxmiss.append(x)
        vymiss.append(y)
    contador += 1
proporcion = hit/contador
pi = 4 * proporcion
print(pi)
````

## Introducción - Simular el lanzamiento de dos monedas

Para el lanzamiento de las monedas ¿podemos utilizar el metodo previo?

````
moneda1 = []
moneda2 = []
iteraciones = int(input('Ingresar la cantidad de iteraciones: '))
for x in range(iteraciones):
    moneda1.append(fResultadoMonedaTexto())
    moneda2.append(fResultadoMonedaTexto())
dosmonedas = []
for k in range(len(moneda2)):
    dosmonedas.append(moneda1[k] + '-' + moneda2[k])
dosmonedasunica = list(dict.fromkeys(dosmonedas))
dosmonedasvalores = [None]*len(dosmonedasunica)
print(dosmonedasunica, dosmonedasvalores)
k, m = 0, 0
#ahora modo super sayayin
for i in dosmonedasunica:
    for j in dosmonedas:
        if i == j:
            m += 1
            dosmonedasvalores[k] = m
    k += 1
    m = 0
print(dosmonedasvalores)
````

````
plt.figure(figsize=(10, 10))
plt.bar(dosmonedasunica, dosmonedasvalores, color='royalblue')
plt.grid(color='gray', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.ylabel('Cantidad Experimentos')
plt.xlabel('Pares de monedas')
plt.title('Lanzamiento de dos monedas para {} experimentos'.format(iteraciones))
for l  in range(len(dosmonedasunica)):
    x = dosmonedasunica[l]
    y = dosmonedasvalores[l]
    plt.text(x, y, str(y))
plt.show()
````

## Introducción - Simular el lanzamiento de un dado

Para el lanzamiento de un dado necesitamos estimar las probabilidades del dado.

````
def fResultadoDadoTexto():
    aleatorio = rnd.random()
    if aleatorio <= 1 / 6:
        return '1'
    elif aleatorio <= 2 / 6:
        return '2'
    elif aleatorio <= 3 / 6:
        return '3'
    elif aleatorio <= 4 / 6:
        return '4'
    elif aleatorio <= 5 / 6:
        return '5'
    else:
        return '6'
print(fResultadoDadoTexto())
````

## Simular un juego de Bingo sin repetir letra y numero generado. (Tarea)
### Bingo 75

Es el Bingo americano, muy jugado en red. Tiene formato de cartón y juego de 24 números sobre 75 (la casilla central no tiene número). Basado en la realización de figuras dentro del cartón se pueden obtener diferentes posibilidades de premios de bingo (líneas, diagonales, cruces, etc...). Estas figuras se llaman “patrones”, y cada juego tiene uno o más patrones que dan premio. La composición del cartón es la siguiente: en la columna “B” se ubican los números entre el 1 y el 15, en la columna “I” los números del 16 al 30, en la “N” los números del 31 al 45, en la “G” del 46 al 60 y finalmente en la columna “O” los números del 61 al 75. Los patrones para premios de Bingo como se ha dicho antes, pueden ser de cualquier forma: en diagonal, línea, cruces, cuadrados, letras... En cada juego se especifica qué patrón da premio.

![Bingo75](https://upload.wikimedia.org/wikipedia/commons/4/47/Cartonbingo.jpg)

El ejercicio consiste en crear un cartón de BINGO o en su defecto 'n' cartones utilizando una función que retorne una matriz.

````
import numpy as np
#Necesitas numpy para crear e iterar matrices
#Debemos crear una funcion para crear n cartones
def fCrearCartonBINGO():
    #primero debemos crear las listas con los numeros para poblar la matriz cartonuda
    B = []
    I = []
    N = []
    G = []
    O = []
    X = []
    #Con varios ciclos, tambien se puede hacer con un solo ciclo y condiciones
    for i in range(1, 16):
        B.append(i)
    for i in range(16, 31):
        I.append(i)
    for i in range(31, 46):
        N.append(i)
    for i in range(46, 61):
        G.append(i)
    for i in range(61, 76):
        O.append(i)
    #ya tengo los valores, ahora debo recorrer una matriz de 6x5 con las letras BINGO en la primera fila
    A = np.zeros((6, 5), dtype='O') # Creamos una matrix de 6x5 de tipo txto

    #Ahora para recorrer la matrix debemos iterar los elementos
    #Escribir la palabra BINGO
    bingo = ['B','I', 'N', 'G', 'O']
    columnas = A.shape[1] # 0 es para filas, 1 para columnas
    filas = A.shape[0]
    for q in range(columnas):
        A[0,q] = bingo[q]

    for i in range(filas):
        for j in range(columnas):
            if i > 0:
                if j == 0: # Estamos en la columna 1, debemos llamar la B
                    dato =  rnd.choice(B)
                    B.remove(dato)
                    A[i,j] = dato
                elif j == 1:
                    dato =  rnd.choice(I)
                    I.remove(dato)
                    A[i,j] = dato
                elif j == 2:
                    dato =  rnd.choice(N)
                    N.remove(dato)
                    A[i,j] = dato
                elif j == 3:
                    dato =  rnd.choice(G)
                    G.remove(dato)
                    A[i,j] = dato
                else:
                    dato =  rnd.choice(O)
                    O.remove(dato)
                    A[i,j] = dato
            if i == 3 and j == 2:
                A[i,j] = '-'
    return A
````

````
#@title Seleccionar la cantidad de cartones a generar
cartones = 10 #@param {type:"slider", min:1, max:10, step:1}
for i in range(cartones):
    print('Carton {}'.format(i+1))
    print(fCrearCartonBINGO())
    print('---------------------------')
````

## Simular una mano de dominó. (Tarea)
### Dominó

El dominó es un juego de mesa en el que se juegan y emplean unas fichas (baldosas) rectangulares, generalmente blancas por la cara y negras por el envés, usualmente hay de diferentes colores, aunque existen muchas variantes. Una de sus caras está dividida por dos cuadrados, cada uno de los cuales está numerado normalmente mediante disposiciones de puntos como los dados.
La puntuación habitual es de cero a seis puntos, lo que compone un total de 28 piezas de dominó siendo la ficha más grande la del seis doble. Existen otras variantes de 55, 91, 136 y 190 piezas. Por ejemplo, en Cuba se juega con 28 piezas en la zona oriental, mientras que en el centro y occidente, incluyendo la capital, se juega con 55 fichas (Doble 9). El jugador que gana sin fichas son 3 puntos y por medio de puntos son 2. Si un jugador al de la partida obtiene cuatro fichas con iguales números en ambos lados es considerado como victorioso y la partida será reiniciada.

![Dominó](https://upload.wikimedia.org/wikipedia/commons/a/ae/Dominospiel.JPG)

````
#Vamos a crear una funcion desde el principio
import random as rnd
def fManoDomino(jugadores): #La cantidad de jugadores determina la cantidad de fichas de 2 a 4 jugadores siempre 7 fichas
    if jugadores >= 2 and jugadores <= 4:
        domino = []
        for i in range(0,7):
            for j in range(0, i+1):
                domino.append('|' + str(i) + '|' + str(j) + '|')
        #ya tengo el domino, debo establecer la salida de las fichas
        #Debemos tener siempre al menos 2 jugadores, realicemos los dos jugadores
        manos = []
        jugador1 = []
        jugador2 = []
        #------------jugador 1----------------
        for i in range(7):
            dato =  rnd.choice(domino)
            domino.remove(dato)
            jugador1.append(dato)
        #------------jugador 2----------------
        manos.append(jugador1)
        for i in range(7):
            dato =  rnd.choice(domino)
            domino.remove(dato)
            jugador2.append(dato)
        manos.append(jugador2)
        if jugadores >= 3:
        #------------jugador 3----------------
            jugador3 = []
            for i in range(7):
                dato =  rnd.choice(domino)
                domino.remove(dato)
                jugador3.append(dato)
            manos.append(jugador3)
            if jugadores >= 4:
                jugador4 = []
            #------------jugador 4----------------
                for i in range(7):
                    dato =  rnd.choice(domino)
                    domino.remove(dato)
                    jugador4.append(dato)
                manos.append(jugador4)
        return manos
    else:
        return 'Solo se permiten entre 2 y 4 jugadores, vuelva a intentarlo'
fManoDomino(4)
````
