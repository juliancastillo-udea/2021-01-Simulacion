#Generador Congruencial

##Funciones necesarias
Generador de numeros primos y generador de semilla basado en tiempo

```python
import matplotlib.pyplot as plt
import statistics as st
from datetime import datetime
import math as mt

#Generador de numeros primos
def NumerosPrimos(inferior,superior):
    primos = []
    for num in range(inferior,superior + 1):  
       if num > 1:  
           for i in range(2,num):  
               if (num % i) == 0:  
                   break  
           else:  
               primos.append(num)
    return primos
#Generador de semilla
def Semilla():
    tiempo = datetime.now()
    milisegundos = float(tiempo.strftime('.%f'))
    milisegundos = round(milisegundos, 6)
    milisegundos = int(milisegundos * 1000000)
    return milisegundos
```

```python
def GeneradorCongruencialMixto(semilla, listaprimos, secuencia=1000): #Por defecto se generaran mil numeros aleatorios
    strSemilla = str(semilla)
    x = []
    r = []
    x.append(semilla)
    #tomo los dos primeros numeros y los dos ultimos para los numeros primos, debemos tener al menos 100 numeros primos generados
    a = int(strSemilla[:2])
    c = int(strSemilla[-2:])
    m = 2**(len(strSemilla))
    for i in range(1, secuencia + 1):
        n = (a*x[i-1]+c)
        n = n % m
        x.append(n)
        r.append((n/m))
    return r
```

```python
y = GeneradorCongruencialMixto(Semilla(), NumerosPrimos(10000,100000))
#print(y)
print(st.mean(y))
print(st.variance(y))
```
