En teoría de números, el [último teorema de Fermat](https://es.wikipedia.org/wiki/%C3%9Altimo_teorema_de_Fermat), o teorema de Fermat-Wiles, es uno de los teoremas más famosos en la historia de las matemáticas. Utilizando la notación moderna, se puede enunciar de la siguiente manera:
Si $n$ es un número entero mayor o igual que $3$, entonces no existen números enteros positivos $x$, $y$ y $z$, tales que se cumpla la igualdad:

$x^n + y^n = z^n$

Este teorema fue conjeturado por Pierre de Fermat en 1637, pero no fue demostrado hasta 1995 por Andrew Wiles ayudado por el matemático Richard Taylor. La búsqueda de una demostración estimuló el desarrollo de la teoría algebraica de números en el siglo XIX y la demostración del teorema de la modularidad en el siglo XX.
